# StandardScaler
- It will make sure that data stay nearby 0

### z = (x - m)/σ
```bash 
x = Actual value
m = mean of the column
σ = Standard Deviation of a column

z = Standardised value (Scaled Output)
```

````python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split

data = {
    "StudyHours": [1,2,3,4,5],
    "TestScore": [40,50,60,70,80]
}

df = pd.DataFrame(data)

standard_scaler = StandardScaler()
standard_scaled = standard_scaler.fit_transform(df)

print("Standard Scaler Output")
print(pd.DataFrame(standard_scaled, columns = ["StudyHours", "TestScore"]))
````

# MinMaxScaler

- It make sure that data stay within 0 and 1

### Scaled = (x - y)/(z - y)
```bash

x = Actual value
y = Smallet value from column
z = Biggest value from column
```
- Example

```bash
[1,2,3,4,5]

(x - y)/(z - y) = Scaled

if x is 1 then (1 - 1)/(5 - 1) = 0.00
if x is 2 then (2 - 1)/(5 - 1) = 0.25
if x is 3 then (3 - 1)/(5 - 1) = 0.50
if x is 4 then (4 - 1)/(5 - 1) = 0.75
if x is 5 then (5 - 1)/(5 - 1) = 1.00
```
---

````python
minmax_scaler = MinMaxScaler()
minmax_scaled = minmax_scaler.fit_transform(df)

print("MinMax Scaler Output")
print(pd.DataFrame(minmax_scaled, columns= ["StudyHours", "TestScore"]))
````
